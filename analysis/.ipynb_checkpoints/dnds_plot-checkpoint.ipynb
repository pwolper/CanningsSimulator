```python
import glob
import re
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
csv_files = glob.glob("../src/simulations/*.csv")
```

```python
df_list = []

for file in csv_files:
    df = pd.read_csv(file)
    df_list.append(df)

data = pd.concat(df_list, ignore_index=True)
data.drop(columns=['selection_period','selection_type', 'population_size_model', 'lost_probability'], inplace=True)
```

```python
data = data.sort_values(by='selection_coefficient')
data
```

```python
data = data[data['N_simulations'] == 250000]
data['NeS'] = data['selection_coefficient'] * data['population_size']
data
```

```python
groups = data.groupby('parameter')
final_data = []
for name, group in groups:
    ds = group.loc[group['selection_coefficient'] == 0, 'fixation_probability'].values
    print('ds = ', ds)
    print('parameter:', name)
    group['dn/ds'] = group['fixation_probability']/ds
    group['log dn/ds'] = np.log(group['dn/ds'] + 1)

    plt.axvline(x=0, color='black')
    plt.axhline(y=1, color='black')
    plt.plot(group['NeS'], group['log dn/ds'], label=name)
    plt.ylabel('log dn/ds')
    plt.xlabel('NeS')
    plt.legend(title=r'$\alpha$')
    
    final_data.append(group)
```

```python
final_data = pd.concat(final_data)
final_data
```
